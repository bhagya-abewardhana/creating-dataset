```
values=[]
with open("Musical.json","r") as f:
    for i in f:
        values.append(json.loads(i))
print(values[:5])
```

```
import csv
total = len(values)
batch = 20522
count = 0

for i in range(0,total,batch):
    print("iter number: ",count/batch)
    review = []
    lable = []
    for j in range(count , count+batch):
        try:
            rtext = values[j]["reviewText"]
            rover = values[j]["overall"]
            
            
            review.append(rtext)
            lable.append(rover)
           
            
            
        except:
            pass
    count = count + batch
    with open("Musicalnew.csv" , "a",newline = "") as csvfile:
        xx = list(zip(lable,review))
        writer = csv.writer(csvfile,delimiter = ",")
        for k in xx:
            writer.writerow(k)
```

```
colnames = ["lable","review"]
df= pd.read_csv("Musicalnew.csv",names= colnames,header = None)
```

```
newdf = df
newdf.loc[1,'lable'] ==3
total = len(newdf)
batch = 20522
pr = 100
count=0
for i in range(0,total-1,batch):
    print("MAIN iteration is: ", count/batch)
    Sentiment = []
    for j in range(count,count+batch):
        if j%pr ==0:
            print("INTER iteration is: ", j/pr)
        vote_val = newdf.loc[count,"lable"]
        if vote_val == 3:
            Sentiment.append("neu")
        elif vote_val>=0 and vote_val < 3:
            Sentiment.append("neg")
        else:
            Sentiment.append("pos")
    print(len(Sentiment))
    print(count)
    print(count+batch)
    newdf.loc[count:count+batch-1,'Sentiment'] = Sentiment
    count = count+batch
```

```
newdf.to_csv("processedData.csv",chunksize=20522)
```
